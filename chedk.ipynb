```python
import inspect
import kcorrect.kcorrect as kk

kc = kk.Kcorrect(responses=["sdss_r0", "sdss_g0"], responses_out=["sdss_r0", "sdss_g0"], responses_map=["sdss_r0", "sdss_g0"])

print("Callable public methods on Kcorrect:")
for name, obj in inspect.getmembers(kc, predicate=callable):
    if not name.startswith("_"):
        print(" -", name)

print("\nSignature hints (common candidates):")
candidates = ["fit_coeffs", "kcorrect", "reconstruct", "reconstruct_maggies", "predict_maggies", "photo", "maggies"]
for n in candidates:
    if hasattr(kc, n):
        print(f"\n{n}: {inspect.signature(getattr(kc, n))}")
```

```python
import numpy as np
from lfkit.corrections.kcorrect_backend import fit_coeffs_from_mag_color, build_kcorrect

coeffs, fit_responses = fit_coeffs_from_mag_color(
    mag_band="sdss_r0",
    mag=20.0,
    color=("sdss_g0", "sdss_r0"),
    color_value=0.85,
    z_phot=0.0,
)

kc = build_kcorrect(responses_in=fit_responses, responses_out=fit_responses, responses_map=fit_responses)
pred = np.asarray(kc.reconstruct(redshift=0.0, coeffs=coeffs), float)

print("fit_responses:", fit_responses)
print("pred maggies:", pred)
print("coeff sum:", coeffs.sum())
```
